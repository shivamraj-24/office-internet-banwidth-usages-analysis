Office Internet Bandwidth Usage Analysis

. From: Minor 1 – Statistics

. Extension: Analyze usage → peak load detection

. Add-ons: Line plots, variance analysis

. Difficulty: Medium


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)

```

```python
hours = list(range(24))
bandwidth = np.random.randint(20, 120, size=24)
```

```python
data = pd.DataFrame({
    "Hour": hours,
    "Bandwidth_Usage_MBPS": bandwidth
})
```

```python
print(data)
```

```python
# Line plot
plt.figure(figsize=(10,5))
plt.plot(data["Hour"], data["Bandwidth_Usage_MBPS"], marker='o')
plt.xlabel("Hour of the Day")
plt.ylabel("Bandwidth Usage (Mbps)")
plt.title("Office Internet Bandwidth Usage Over Time")
plt.grid(True)
plt.show()

```

```python
# Peak detection
peak_hour = data.loc[data["Bandwidth_Usage_MBPS"].idxmax()]
print("\nPeak Load Hour:")
print(peak_hour)
```

```python
# Variance analysis
variance = data["Bandwidth_Usage_MBPS"].var()
mean_usage = data["Bandwidth_Usage_MBPS"].mean()
```

```python
print("\nMean Bandwidth Usage:", mean_usage)
print("Variance of Usage:", variance)

```

```python
# Highlight peak
plt.figure(figsize=(10,5))
plt.plot(data["Hour"], data["Bandwidth_Usage_MBPS"], marker='o', label="Usage")
plt.scatter(peak_hour["Hour"], peak_hour["Bandwidth_Usage_MBPS"], s=150, label="Peak Load")
plt.xlabel("Hour of the Day")
plt.ylabel("Bandwidth Usage (Mbps)")
plt.title("Peak Load Detection in Office Internet Usage")
plt.legend()
plt.grid(True)
plt.show()
```
